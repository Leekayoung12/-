```python
# 과제
# 지니 차트 1위~200위까지 크롤링
# 반복문 사용하여 크롤링
# https://www.genie.co.kr/chart/top200
# https://www.genie.co.kr/chart/top200?ditc=D&ymd=20231122&hh=14&rtm=Y&pg=1
# https://www.genie.co.kr/chart/top200?ditc=D&ymd=20231122&hh=14&rtm=Y&pg=2
# https://www.genie.co.kr/chart/top200?ditc=D&ymd=20231122&hh=14&rtm=Y&pg=3
# https://www.genie.co.kr/chart/top200?ditc=D&ymd=20231122&hh=14&rtm=Y&pg=4
# 엑셀로 내보내기 기능
# 과제 제출은 깃허브로 내보내기
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
site = 'https://www.genie.co.kr/chart/top200'
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36'}
dic = {}
dic_list = []

# 4페이지에 걸쳐 크롤링
for page in range(1, 5):
    # 페이지별 URL 조합
    page_url = f'{site}?ditc=D&ymd=20231122&hh=14&rtm=Y&pg={page}'

    # 페이지 요청 및 HTML 파싱
    request = requests.get(page_url, headers=header)
    soup = BeautifulSoup(request.text, 'html.parser')

    # 'td' 태그 중 'class'가 'info'인 것을 찾아서 정보 수집
    chart = soup.findAll('td', {'class': 'info'})

    # 반복문을 통해 각 'td' 태그 안의 정보 추출
    for i, td in enumerate(chart, start=((page - 1) * 50) + 1):
        # 곡 제목 추출
        title_element = td.find('a', {'class': 'title ellipsis'})
        if title_element:
            title = title_element.text.strip().replace('[19금]\n ','')
        else:
            title = "Unknown Title"

        # 아티스트 추출
        artist_element = td.find('a', {'class': 'artist ellipsis'})
        if artist_element:
            artist = artist_element.text.strip()
        else:
            artist = "Unknown Artist"

        # print(f"{i}위 {title} - {artist}")
        dic = {'num': i, 'title': title, 'artist': artist}
        dic_list.append(dic)

print(dic_list)

```
